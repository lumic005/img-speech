```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread(r"C:\Users\shari\Downloads\g.jpg")
median = cv2.medianBlur(img, 5)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(1, 2, 2)
plt.imshow(median)
plt.title('Median Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```
