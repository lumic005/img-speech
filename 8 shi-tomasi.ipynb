```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread(r"C:\Users\shari\Downloads\g.jpg")
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
corners = cv.goodFeaturesToTrack(gray, maxCorners=100, qualityLevel=0.01, minDistance=10)
corners = np.int0(corners)
shi_img = img.copy()
for i in corners:
    x, y = i.ravel()
    cv.circle(shi_img, (x, y), 3, (0, 255, 0), -1)
plt.subplot(121)
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")
plt.subplot(122)
plt.imshow(cv.cvtColor(shi_img, cv.COLOR_BGR2RGB))
plt.title("Shi-Tomasi Corners")
plt.axis("off")
plt.show()
```
